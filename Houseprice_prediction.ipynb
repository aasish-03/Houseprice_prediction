    #Import All the required packages







```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
#Read the dataset
df=pd.read_csv('Housing.csv')
df.head(5) #Display the first 5 rows of dataset
```

```python
#Checking for Null values
df.isnull().sum()
#Identify the Categorical features in the dataset as they need to be encoded before fitting the model
col=df.select_dtypes(include='object').columns
cols=list(col)
```

```python
#Create dummy variables for Categorical features as part of encoding and combine them with original dataset
dummy=[]
dummy=pd.get_dummies(df[cols],columns=cols)
dummies=pd.DataFrame(dummy)
data=[df,dummies]
newdf=pd.concat(data,axis=1)
newdf.info()
```

```python
#Drop all the less significant columns/features from the dataset
list1=["mainroad_no","guestroom_no","basement_no","hotwaterheating_no","airconditioning_no","prefarea_no","mainroad","guestroom","basement","hotwaterheating","airconditioning","prefarea","furnishingstatus"]
newdf=newdf.drop(newdf[list1],axis=1)
newdf.info()
```

```python
#Plot a 2D-heatmap to check the correlation of attributes with target variable(price)
data=newdf.corr()
plt.figure(dpi=200)
sns.heatmap(data,annot=True,fmt='.2f')
plt.show()
```

```python
#From the heatmap it is observed that features (hotwaterheating_yes,furnishingstatus_semi-furnished) are less correlated with price variable
#Hence we can drop them
newdf=newdf.drop(newdf[["hotwaterheating_yes","furnishingstatus_semi-furnished"]],axis=1)
newdf.head()
```

   #Splitting the dataset into train and test datasets

```python
X=newdf.values
y=newdf['price']
X=np.delete(X,0,1)
y=np.array(y)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=50)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
y_test.shape
```

  #Fitting LinearRegression model on the training dataset

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
#Predicting the target variable values on train data
pred=model.predict(X_train)
pred
#Calculating the r^2 metric for predictions and actual target values on test dataset
r_train=r2_score(y_train,pred)
```

```python
#Predicting the target variable values on test data
pred=model.predict(X_test)
pred
#Calculating the r^2 metric for predictions and actual target values on test dataset
r_test=r2_score(y_test,pred)
print(r_train,r_test)
```
